```python
import requests
from bs4 import BeautifulSoup

LIMIT = 50
URL = f"https://www.indeed.com/jobs?q=python&limit={LIMIT}"

def extract_indeed_pages():
    result = requests.get(URL)
    soup = BeautifulSoup(result.text,'html.parser')

    pagination = soup.find("ul","pagination-list")

    #ul태그의 pagination-list클래스에 해당하는 soup형를 반환하여
    #a태그를 가진 것 모두를 리스트 형태로 저장함
    pages = pagination.findAll('a') # == .find_all

    spans=[]
    #필요 없는 맨 마지막 값 제거
    for page in pages[:-1]:
        spans.append(int(page.string))

    #리스트의 맨 마지막 값(마지막 페이지) 대입후 반환
    last_page = spans[-1]

    return last_page

def extract_indeed_jobs(last_page):
    #페이지에 따라 주소에 &start=50*(page-1)이 추가됨
    #따라서 0(0*50), 50(1*50), 100(2*50) .. 순으로 start에 넣어줌
    for page in range(last_page):
        result = requests.get(f"{URL}&start={page*LIMIT}")
        soup = BeautifulSoup(result.text,"html.parser")
        
        results = soup.findAll("a","tapItem")
        for result in results:
            title = result.find('h2','jobTitle')
            #span태그에 title이 없는 경우 필요없는 값이 포함되어있기 때문에
            #title이 있는 span 태그 찾기
            print(title.find('span',title=True).string)
    

last_indeed_page = extract_indeed_pages() #5
extract_indeed_jobs(last_indeed_page)
```
