```python
lunch = ["된장찌개", "피자", "제육볶음", "짜장면"]

#집합으로 변환
set_lunch = set(lunch)
print(set_lunch)

#요소 추가
set_lunch.add("라면")
print(set_lunch)

#요소 삭제
set_lunch.remove("피자")
print(set_lunch)

#요소 여러개 추가. 단, 이렇게 쓰면 김, 치가 따로 입력됨
set_lunch.update("김치")
print(set_lunch)

#리스트로 묶어서 추가하기
set_lunch.remove("치")
set_lunch.remove("김")
set_lunch.update(["김치","된장찌개"])
print(set_lunch)

# 집합 간의 연산 ->> 합집합: |, 차집합: -, 교집합: &
```

```python
def make_incrementor(n):
    print("my n is :",n)
    return lambda x: x+n

inc_42f = make_incrementor(42)
inc_13f = make_incrementor(13)
print(inc_42f(0))
print(inc_42f(4))
print(inc_13f(0))
print(inc_13f(4))

#튜플 정렬
pairs=[(2,'two'),(1,'one'),(3,'three'),(4,'four')]
print(pairs)
#첫 번째 원소로 정렬
pairs.sort(key=lambda pair:pair[0])
print(pairs)
#두 번째 원소로 정렬
pairs.sort(key=lambda pair:pair[1])
print(pairs)
```

```python
def toList(str):
    print(list(str))
    return list(str)

def toDict(list):
    dic={}
    for i in list:
        if i in dic:
            dic[i]+=1
        else:
            dic[i]=1
    print(dic)
    return dic

def toMap(dict):
    for key,value in dict.items():
        print('Convert to Map! {0} {1}'.format(key,value))

a=input("입력:")
convertedList=toList(a)
convertedDict=toDict(convertedList)
toMap(convertedDict)
```

```python
str_test1 = "stringtest"
str_test2 = "stringtest"
int_test1 = 3
int_test2 = 3
list_test1 = [1,2]
list_test2 = [1,2]

#str: 같은 객체 (띄어쓰기 포함된 스트링은 제외)
#단, str은 값이 같다고 항상 같은 값을 참조하는 것은 아님.
#따라서 == 비교 연산을 사용하는 것이 맞음
#immutable
print(str_test1 == str_test2)
print(str_test1 is str_test2)

#int: 같은 객체
#immutable
print(int_test1 == int_test2)
print(int_test1 is int_test2)

#list: 값은 같지만 다른 객체
#mutable
print(list_test1 == list_test1)
print(list_test1 is list_test2)

if type(str_test1) == str:
    print(f"hi {type(str_test1)}")
```
