```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime

url = "https://www.daum.net"

#일부 웹사이트의 경우 크롤링을 막는 경우가 있음. -> headers 포함해야함
response = requests.get(url)

# 해당 주소에 있는 HTML 코드 가져옴 (str 형태)
response.text 
# response.encoding : 해당 주소에 있는 인코딩 방식 가져옴 (str 형태)
# 이외 다양한 정보 불러오기 가능

# BeautifulSoup(데이터, 파싱 방법)
# 해당 주소에 있는 HTML 코드를 가져옴 (BeautifulSoup 형태)

soup = BeautifulSoup(response.text,'html.parser')

#soup.title -> <title>Daum</title>
#soup.title.string -> Daum
#soup.span -> <span ~~> ~~ </span> -> 맨 처음 span만 가져옴
#soup.findAll('span') -> 모든 span 태그에 대한 값 가져옴

print(datetime.today().strftime("%Y년 %m월 %d일의 실시간 검색어 순위입니다.\n"))

#a태그를 가진 것 중에 link_favorsch 클래스를 가진 것들만 리스트 형태로 가져옴
#단, 태그와 함께 가져옴, 각 원소들은 bs4.element.Tag 타입
results = soup.findAll('a','link_favorsch')

for result in results:
    #태그를 제외한 값을 가져옴
    print(result.get_text())


#파일 쓰기 -> /content/daum.html 로 저장
File = open('daum.html','w')
File.write(response.text) #str만 입력 가능
File.close()
```

```python
import requests
from bs4 import BeautifulSoup as bs
from datetime import datetime

url = "https://www.nate.com/"


response = requests.get(url)

soup = bs(response.text,'html.parser')

title = soup.title
#span 태그의 txt_rank 클래스를 모두 가져옴
results = soup.findAll('span', "txt_rank")


File = open('nate_ranking.txt','w')

rank = 1
today =datetime.today().strftime("%Y년 %m월 %d일의 실시간 검색어 순위입니다.\n") 

File.write(today+'\n')
print(today)
for result in results:
    print(rank, f'위 {result.get_text()}') #result.get_text() == result.string
    File.write(str(rank)+'위 '+result.get_text()+'\n')
    rank+=1

File.close()
```
