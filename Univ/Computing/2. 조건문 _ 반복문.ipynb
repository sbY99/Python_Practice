```python
# 실수 비교
from math import sqrt 
n = sqrt(3.0)

# n은 3.0을 루트화 하면서
# 소수점 자리에서 잘리기 때문에, 정확한 루트3이 아님
# 따라서 제곱했을 때 3.0이 나오지 않음
if n*n == 3.0 :
    print("sqrt(3.0)*sqrt(3.0)은 3.0과 같다.") 
else :
    print("sqrt(3.0)*sqrt(3.0)은 3.0과 같지 않다.")

# 따라서, n의 제곱과 3.0의 차이가
# 매우 작으면, 같다고 판단해줌
if abs(n*n - 3.0) < 0.00001 :
    print("sqrt(3.0)*sqrt(3.0)은  3.0과  같다. ")
```

```python
# Lab: 방정식의 해 구하기 (f(x) = x^2 -x -1)
count = 100
start = 1.0
end = 2.0

for i in range(count):
    x = start + i*(end-start)/count
    f = (x**2-x-1)
    if abs(f-0.0) < 0.01:
        print(f'방정식의 해는 {x}')
```

```python
# Lab: 숫자 맞추기 게임
import random

tries = 0
response = 0
answer = random.randint(1,100)
print('1부터 100사이의 숫자를 맞추시오')

while response!=answer:
    response = int(input('숫자를 입력하시오:'))
    tries+=1
    if response < answer:
        print('너무 낮음!')
    elif response>answer:
        print('너무 높음!')
print(f'축하합니다. 시도횟수={tries}')
```

```python
# Lab: 파이 계산하기
count = int(input('반복횟수: '))
result = 0
idx=0

while(count > 0):
    divisor = idx*2+1
    if(idx%2==0):
        result+=4/divisor
    else:
        result-=4/divisor
    idx+=1
    count-=1
print(f'Pi ={result}')
```

```python
# Lab: 도박상의 확률
import random

initMoney = 50
goalMoney = 250

count=0
success=0
while count<100:
    curr = initMoney
    while curr>0 and curr <goalMoney:
        r = random.randint(0,1)
        if r==0:
            curr+=1
        else:
            curr-=1
    if curr == goalMoney:
        success+=1

    count+=1
print(f'100번 중에서 {success}번 성공')
```
