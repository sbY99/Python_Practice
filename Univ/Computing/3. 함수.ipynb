```python
# 가변인수 예제
def add(*numbers):
    sum = 0
    for n in numbers:
        sum = sum + n
    return sum

print(add(10,20))
print(add(10,20,30))
```

```python
# 매개 변수 이름 앞에 이중별표를 사용하여
# 가변 길이 키워드 인수를 나타냄
# 인수는 딕셔너리 형태로 전달
def myfunc(**kwargs):
    print(kwargs)
    result = ''
    for arg in kwargs.values():
        result+=arg
    return result

print(myfunc(ace='Hi!',bee='Mr.', name='Kim'))
```

```python
# Lab: 주급 계산 프로그램
def weeklyPay(rate, hour):
    if(hour>30):
        return (rate*30 + 1.5*rate*(hour-30))
    else:
        return rate*hour

rate = int(input('시급을 입력하시오:'))
hour = int(input('근무 시간을 입력하시오:'))
print(f'주급은 {weeklyPay(rate,hour)}')
```

```python
# 지역변수 vs 전역변수
gx = 100 # 전역변수
def myfunc():
    x=100 # 지역변수
    print(x)
    print(gx)
myfunc()
```

```python
# 함수 안에서 전역 변수 변경하기
gx = 100
g = 200
def myfunc():
    gx = g #함수 안에서 변수에 값을 저장하면 지역 변수가 생성됨
    print(gx)

myfunc()
print(gx) # 전역변수는 변경되지 않았음
```

```python
# 함수 안에서 전역 변수 변경하기
gx = 100
g = 200
def myfunc():
    global gx # 전역변수 gx를 사용할 것임을 명시
    gx = g 
    print(gx)

myfunc()
print(gx) # 전역변수가 변경됨
```

```python
# 실습 문제2
def minValue(a, b, c):
    if a<b and a<c:
        return a
    elif b<a and b<c:
        return b
    else:
        return c

a,b,c = map(int, input('3개의 정수를 입력하시오:').split(','))
print(f'제일 작은 정수는 {minValue(a,b,c)} 입니다.')
```

```python
# 실습 문제3
import math
def get_value(a,b,c):
    v1 = (-b + math.sqrt(b**2.0-4.0*a*c))/(2.0*a)
    v2 = (-b - math.sqrt(b**2.0-4.0*a*c))/(2.0*a)
    
    return v1,v2

a = int(input('a를 입력하시오: '))
b = int(input('b를 입력하시오: '))
c = int(input('c를 입력하시오: '))
v1, v2 = get_value(a,b,c)
print(f'두 실근은 {v1}, {v2}입니다.')
```
