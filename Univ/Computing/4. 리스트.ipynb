```python
# Lab: 전치 행렬 계산
transpose = []
matrix = [[1,4,7],[2,5,8],[3,6,9]]

print('원래 행렬=',matrix)

# 열의 개수만큼 반복
for i in range(len(matrix[0])):
    transposed_row =[]
    for row in matrix:
        transposed_row.append(row[i])
    transpose.append(transposed_row)
print('전치 행렬=',transpose)
```

```python
# Lab: Tic-Tac-Toe
board = [[' ' for x in range(3)]for y in range(3)]

while True:
    for r in range(3):
        print('  '+board[r][0] + '|  '+board[r][1] + '|  '+board[r][2])
        if (r<=2):
            print('---|---|---')
        
    x = int(input('다음 수의 x좌표를 입력하시오:'))
    y = int(input('다음 수의 y좌표를 입력하시오:'))
    
    if board[y][x]!=' ':
        print('잘못된 위치입니다.')
        continue
    else:
        board[y][x]='X'
    
    done=False
    for i in range(3):
        for j in range(3):
            if board[i][j] ==' ' and (not done):
                board[i][j] = 'O'
                done=True
                break

    isFinished = 0
    for i in range(3):
        for j in range(3):
            if board[i][j] !=' ':
                isFinished+=1
    if isFinished == 9:
        break
```

```python
# 실습문제 2
n = int(input('몇 번째 항까지 구할까요? '))
def fibo(n):
    if n==0:
        return 0
    if n==1:
        return 1
    else:
        return fibo(n-1) + fibo(n-2)    

for i in range(n+1):
    if i < n:
        print(f'{fibo(i)}, ',end='')
    else:
        print(fibo(i))
```

```python
# 실습문제 6
def dec2bin(dec):
    lst=[]
    while(dec!=0):
        lst.append(dec%2)
        dec//=2
    bin=''
    for i in lst[::-1]:
        bin += str(i)
    print(bin)
    
n = int(input('10진수: '))
dec2bin(n)
```
